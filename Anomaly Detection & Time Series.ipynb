```python
# 1. What is Anomaly Detection? Explain its types (point, contextual, and
# collective anomalies) with examples.
"""
Anomaly Detection is the task of identifying unusual patterns, rare events, or data points that do not follow the expected behavior.
These unusual points are called anomalies or outliers.

Types of anomalies:
1. Point Anomaly : 
    - A single data point that is significantly different from the rest of the dataset.
    - If most credit card transactions are between ₹200–₹3,000 but one is ₹90,000 → this is a point anomaly.
    
2. Contextual Anomaly : 
    - A data point that is normal in general but becomes abnormal in a specific context (such as time, season, or location).
    - A temperature of 32°C is normal during the afternoon but abnormal at 3 AM, making it a contextual anomaly.

3. Collective Anomaly : 
    - A group or sequence of data points that are abnormal together, even if individual points are normal.
    - Receiving 200 network requests per second for 50 consecutive seconds indicates a DDoS attack
"""
```

```python
# 2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
# their approach and suitable use cases
"""
1. Isolation Forest
    -Uses random decision trees to isolate points.
    -Points that get isolated quickly (require fewer splits) are marked as outliers.
    
    - Best for high-dimensional datasets (10–1000+ features).
    - Works well for detecting global outliers far from the main data.
    
2. DBSCAN
    - A density-based algorithm: forms clusters where points are close and dense.
    - Points that do not belong to any dense region are labeled as noise.
    
    - Good for spatial / 2D / 3D data where cluster shape matters.
    - Best for finding outliers around cluster boundaries and global isolated points.
    
3. Local Outlier Factor
    - Measures local density around each point.
    - Points whose density is much lower than neighbors are marked as local anomalies.
    
    - Best when outliers are near clusters, not far away.
    - Excellent for detecting subtle, density-based anomalies like borderline or contextual outliers.
"""
```

```python
# 3. What are the key components of a Time Series? Explain each with one
# example.
"""
1. Trend : 
    - A long-term increase or decrease in the data over time. It represents the general direction.
    - The number of internet users increasing every year shows an upward trend.

2. Seasonality: 
    - Regular and repeating patterns that occur at fixed intervals (daily, weekly, monthly, yearly).
    - Ice cream sales peak every summer and drop during winter → strong seasonality.

3. Cyclic Patterns : 
    - Long-term oscillations that do not follow a fixed period (unlike seasonality). Often driven by economic or business cycles.
    - Stock market ups and downs due to economic cycles (boom → recession → recovery).

4. Random or Irregular Component (Noise) : 
    - Unexpected variations caused by random and unpredictable events.
    - Sudden drop in airline passengers due to a temporary storm → random noise.

5. Level : 
    - The baseline average value around which the time series fluctuates.
    - A company consistently makes around 500 sales per day—this 500 is the level.

6. Stationarity : 
    - A property where mean, variance, and covariance do not change over time.
    -White noise data that fluctuates around a constant mean (0) with constant variance.
"""
```

```python
# 5.  Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
# terms of structure and application.
"""
1. AR (Autoregressive Model):
    - Uses past values of the variable to predict future values.
    - Best when the series depends strongly on its own past observations.
    - Example: Forecasting stock prices using previous day prices.

2. MA (Moving Average Model):
    - Uses past error terms (residuals) to predict future values.
    - Best when random shocks or sudden fluctuations affect the series.
    - Example: Forecasting sales affected by random external spikes.

3. ARIMA (Autoregressive Integrated Moving Average):
    - Combines AR + differencing (I) + MA components.
    - Used for non-stationary, non-seasonal series with trend.
    - Example: Predicting monthly sales that show an increasing trend.

4. SARIMA (Seasonal ARIMA):
    - Extension of ARIMA that adds seasonal AR, differencing, and MA terms.
    - Best for data with clear seasonal patterns (daily/weekly/yearly).
    - Example: Retail sales increasing every December due to festivals.

5. SARIMAX (Seasonal ARIMA with Exogenous Variables):
    - SARIMA + external features (X) such as weather, marketing, holidays.
    - Used when external variables influence the time series.
    - Example: Electricity demand affected by temperature or marketing spend.

"""
```

```python
# 6. Load a time series dataset (e.g., AirPassengers), plot the original series,
# and decompose it into trend, seasonality, and residual components
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

df = pd.read_csv("AirPassengers.csv")
df.rename(columns={"#Passengers": "Passengers"}, inplace=True)
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

plt.figure(figsize=(12, 5))
plt.plot(df['Passengers'])
plt.title("AirPassengers - Original Time Series")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.grid(True)
plt.show()

result = seasonal_decompose(df['Passengers'], model='multiplicative', period=12)
result.plot()
plt.show()
```

```python
# 7.  Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
# detect anomalies. Visualize the anomalies on a 2D scatter plot.


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

df = pd.read_csv("AirPassengers.csv")
df.rename(columns={"#Passengers": "Passengers"}, inplace=True)
df["Month"] = pd.to_datetime(df["Month"])
df.set_index("Month", inplace=True)

df["Month_Number"] = np.arange(len(df))
X = df[["Month_Number", "Passengers"]].values

model = IsolationForest(contamination=0.05, random_state=42)
df["anomaly"] = model.fit_predict(X)

inliers = df[df["anomaly"] == 1]
outliers = df[df["anomaly"] == -1]

plt.figure(figsize=(10, 6))
plt.scatter(inliers["Month_Number"], inliers["Passengers"], s=25, label="Normal Points")
plt.scatter(outliers["Month_Number"], outliers["Passengers"], s=60, color="red", label="Anomalies")
plt.title("Isolation Forest Anomaly Detection - AirPassengers Dataset")
plt.xlabel("Time (Numeric Index)")
plt.ylabel("Passengers")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 8.  Train a SARIMA model on the monthly airline passengers dataset.
# Forecast the next 12 months and visualize the results.

import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv("AirPassengers.csv")
df.rename(columns={"#Passengers": "Passengers"}, inplace=True)
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

model = sm.tsa.statespace.SARIMAX(
    df['Passengers'],
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)

result = model.fit()

forecast = result.get_forecast(steps=12)
forecast_values = forecast.predicted_mean
conf_int = forecast.conf_int()

plt.figure(figsize=(12,6))
plt.plot(df['Passengers'], label="Original", linewidth=2)
plt.plot(forecast_values, label="Forecast (Next 12 Months)", color='red')
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], alpha=0.2, color='red')

plt.title("SARIMA Forecast for Next 12 Months")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
# anomalies and visualize them using matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

np.random.seed(42)
cluster = 0.3 * np.random.randn(300, 2) + np.array([2, 2])
global_outliers = np.random.uniform(low=-3, high=6, size=(10, 2))
X = np.vstack([cluster, global_outliers])

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.03)
y_pred = lof.fit_predict(X)

inliers = X[y_pred == 1]
outliers = X[y_pred == -1]

plt.figure(figsize=(8, 6))
plt.scatter(inliers[:, 0], inliers[:, 1], alpha=0.6, s=25)
plt.scatter(outliers[:, 0], outliers[:, 1], color='red', s=40)
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
