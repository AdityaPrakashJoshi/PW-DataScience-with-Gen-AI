```python
# 1. What is hypothesis testing in statistics?
"""
Hypothesis testing is a way to check if a claim about a population is likely to be true by looking at sample data.
We start with a null hypothesis (H0), which is the default assumption, and compare it against an alternative hypothesis (H1), which is what we suspect might actually be true.
By analyzing the data, we decide whether there’s enough evidence to reject the default assumption in favor of the alternative.
"""
```

```python
# 2. What is the null hypothesis, and how does it differ from the alternative
# hypothesis?
"""
The null hypothesis (H0) is the default assumption that nothing has changed or there’s no difference.
The alternative hypothesis (H1) is the opposite — it says there is a change or difference.
In testing, we start by assuming H₀ is true and only reject it if the evidence strongly supports H₁
"""
```

```python
# 3. : Explain the significance level in hypothesis testing and its role in deciding
# the outcome of a test.
"""
The significance level, is the threshold for deciding whether the results are statistically significant.
It represents the probability of rejecting the null hypothesis (H0).
"""
```

```python
# 4. : What are Type I and Type II errors? Give examples of each.
"""
Type I Error (False Positive) Occurs when you reject the null hypothesis even though it is actually true.
Type II Error (False Negative) Occurs when you fail to reject the null hypothesis even though it is false.
"""
```

```python
# 5. What is the difference between a Z-test and a T-test? Explain when to use
# each.
"""
T-test : 
Used to test hypotheses about population means when:
The population standard deviation is unknown and the sample size is small (n ≤ 30)

Z-test:
Used to test hypotheses about population means when:
The population standard deviation is known or the sample size is large (n > 30)
"""
```

```python
# 6. Write a Python program to generate a binomial distribution with n=10 and
# p=0.5, then plot its histogram.

import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5
size = 1000

data = np.random.binomial(n, p, size)

plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(range(n+1))
plt.show()
```

```python
# 7. : Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
# 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

"""Assume : 
    Null hypothesis (H₀): Population mean = 50

    Alternative hypothesis (H₁): Population mean ≠ 50

    Assume population standard deviation σ = 1"""

import numpy as np
from scipy import stats


sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


pop_mean = 50
pop_std = 1
n = len(sample_data)


sample_mean = np.mean(sample_data)


z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))


p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", round(sample_mean, 3))
print("Z-statistic:", round(z_stat, 3))
print("P-value:", round(p_value, 4))


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0).")
else:
    print("Fail to reject the null hypothesis (H0).")
```

```python
# 8. : Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mu = 50
sigma = 5
n = 100

data = np.random.normal(mu, sigma, n)

sample_mean = np.mean(data)
se = stats.sem(data)

ci = stats.t.interval(0.95, df=n-1, loc=sample_mean, scale=se)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label='95% CI')
plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2)
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# 10.  Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.


import numpy as np
import matplotlib.pyplot as plt

def z_score_analysis(data):
    data = np.array(data)
    
    
    mean = np.mean(data)
    std = np.std(data, ddof=0) 
    z_scores = (data - mean) / std
    
    
    print(f"Mean of data: {mean:.2f}")
    print(f"Standard Deviation of data: {std:.2f}")
    print(f"Z-scores:\n{z_scores}")
    
    
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
    plt.title('Histogram of Z-scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
    
    return z_scores

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=5, size=100)

z_scores = z_score_analysis(sample_data)
```
